Video Game Analysis


Data Prepocessing

```python
import numpy as np
import pandas as pd

gameBehavior_df = pd.read_csv("online_gaming_behavior_dataset.csv", sep=',', decimal='.')
gameBehavior_df_cleaned = gameBehavior_df.drop_duplicates()
gameBehavior_df_cleaned = gameBehavior_df_cleaned.dropna(subset=['Age', 'Location', 'GameGenre'])
```

Graphs

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

gameBehaviorMale = gameBehavior_df_cleaned[gameBehavior_df_cleaned["Gender"] == "Male"]
gameBehaviorFemale = gameBehavior_df_cleaned[gameBehavior_df_cleaned["Gender"] == "Female"]
gameBehaviorMale = gameBehaviorMale.groupby("GameGenre")["Age"].mean()
gameBehaviorFemale = gameBehaviorFemale.groupby("GameGenre")["Age"].mean()

fig, ax = plt.subplots()
w, x = 0.4, np.arange(len(gameBehaviorMale.index))
ax.bar(x - w/2, gameBehaviorMale.values, width = w, label='Male')
ax.bar(x + w/2, gameBehaviorFemale.values, width = w, label='Female')
ax.set_xticks(x)
ax.set_xticklabels(gameBehaviorMale.index)
ax.set_ylabel('Values')
ax.set_title('Average Play Time Hours per genre')
ax.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
gameBehavior_df = pd.read_csv("online_gaming_behavior_dataset.csv", sep=',', decimal='.')
gameBehavior_df_cleaned = gameBehavior_df.drop_duplicates()
gameBehavior_df_cleaned = gameBehavior_df_cleaned.dropna(subset=['Age', 'Location', 'GameGenre'])
gameBehavior_df_cleaned = gameBehavior_df_cleaned.groupby("GameGenre")

fig, ax = plt.subplots()
w, x = 0.4, np.arange(len(gameBehavior_df_cleaned["GameGenre"]))
ax.bar(x - w/2, len(gameBehavior_df_cleaned[gameBehavior_df_cleaned["EngagementLevel"] == "High"]), width = w/2, label='High')
ax.bar(x, gameBehavior_df_cleaned[gameBehavior_df_cleaned["EngagementLevel"] == "High"].count, width = w/2, label='Medium')
ax.bar(x + w/2, gameBehavior_df_cleaned[gameBehavior_df_cleaned["EngagementLevel"] == "High"].count, width = w/2, label='Low')
ax.set_xticks(x)
ax.set_xticklabels(gameBehavior_df_cleaned.index)
ax.set_ylabel('Values')
ax.set_title('Average Play Time Hours per genre')
ax.legend()
plt.show()
```
