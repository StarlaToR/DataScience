Data Prepocessing

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt   

def mean(val):
    numbers = val.split('-')                
    return (float(numbers[0].strip()) + float(numbers[1].strip())) * 0.5      

mobile_games = pd.read_csv("../Dataset/mobile-games.csv", sep=',', decimal='.')
mobile_games_cleaned = mobile_games.drop_duplicates()

#clean genres collumns

mobile_games_cleaned["Genre(s)"] = mobile_games_cleaned["Genre(s)"].astype(str).str.replace("'", "", regex=False).str.replace('"', "", regex=False)
mobile_games_cleaned["Genre(s)"] = mobile_games_cleaned["Genre(s)"].str.split("/ ")
mobile_games_exploded = mobile_games_cleaned.explode("Genre(s)")

```

Graphs

```python
salesPerGenre = mobile_games_exploded.groupby("Genre(s)")["Revenue"].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.bar(salesPerGenre.index, salesPerGenre.values, color = 'green')
plt.xticks(rotation = 75)
plt.ylabel("Revenues")
plt.xlabel("Genres")
plt.title("Top 10 highest grossing genre on mobile")
plt.tight_layout()
plt.show()
```

```python
biggestSales = mobile_games_exploded.groupby("Game")["Revenue"].mean().sort_values(ascending=False).head(10)

print(biggestSales)
plt.figure(figsize=(10,6))
plt.bar(biggestSales.index, biggestSales.values)
plt.xlabel("Revenues")
plt.xticks(rotation = 75)
plt.title("Top 10 highest games on mobile")
plt.tight_layout()
plt.show()
```
