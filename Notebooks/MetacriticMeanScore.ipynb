Data Prepocessing

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

metacritic = pd.read_csv("../Dataset/dataset_metacritic_scraper_2025-02-15.csv", sep=',', decimal='.')
metacritic_cleaned = metacritic.drop_duplicates()
metacritic_cleaned = metacritic_cleaned.dropna(subset=['genres', 'metascore'])
metacritic_cleaned['metascore'] = pd.to_numeric(metacritic_cleaned['metascore'], errors='coerce')
metacritic_cleaned['userscore'] = pd.to_numeric(metacritic_cleaned['userscore'], errors='coerce')
```

Graphs

```python
HighFrequencyGenres = metacritic_cleaned["genres"].value_counts().head(20).index
HighFrequencyGenres = metacritic_cleaned[metacritic_cleaned["genres"].isin(HighFrequencyGenres)]
MeanUserScorePerGenre = HighFrequencyGenres.groupby("genres")["metascore"].mean().sort_values(ascending=False).head(10)

plt.bar(MeanUserScorePerGenre.index, MeanUserScorePerGenre.values, color = 'skyblue')
plt.xticks(rotation=90)
plt.xlabel("Genres")
plt.ylabel("Mean Metacritic Score")
plt.title("Top 10 mean Metacritic score per genre")
```
